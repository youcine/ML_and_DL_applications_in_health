# ***Get DATAFRAME ('Tableau') from file using pandas where file is csv format***

```python
import pandas as pd
data = pd.read_csv('DragonMSI-IC50-All.txt',delimiter='\t')
```

```python
print(data)
```

# ***Show the loaded data in table format***

```python
data.head()
```

# ***Load using .iloc[rows,columns] the targeted columns and rows, and convert it to vectors using .values***

```python
x = data.iloc[:,9:]
```

```python
print(x)
```

```python
y = data.iloc[:,2]
```

```python
print(y)
```

# ***Data processing***

# **Normalization **

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler object
scaler = StandardScaler()

# Transform the data
x = scaler.fit_transform(x)
```

# ***Construire l'ensemble de teste et l'ensemble d'apprentissage***

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,shuffle=True)
```

# ***Training the ANN model***

```python
import tensorflow as tf
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(638,))
model.add(tf.keras.layers.Dropout(0.6))
model.add(tf.keras.layers.Dense(1,activation='linear'))
model.compile(loss='mse',optimizer='ADAM')
model.summary()
```

```python
model.fit(X_train,y_train,epochs=500,batch_size=40)
```

```python
print(X_test.shape)
```

# ***Evaluation du modèle en utilisant la partie Teste***

```python
model.evaluate(X_test,y_test)
```

# *** Modéle MLP***

```python
pip install tensorflow-addons
```

```python
import tensorflow as tf
from tensorflow_addons.metrics import RSquare
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(638,))
model.add(tf.keras.layers.Dense(1024,activation='relu'))
model.add(tf.keras.layers.Dropout(0.6))
model.add(tf.keras.layers.Dense(1,activation='linear'))
model.compile(loss='mse',optimizer='ADAM',metrics=['mae',RSquare()])
model.summary()
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
print(model.predict(X_test))
```

# ***Modéle Convolutional Neural Network***

```python
import tensorflow as tf
from tensorflow_addons.metrics import RSquare
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(638,))
model.add(tf.keras.layers.Reshape((638,1)))
model.add(tf.keras.layers.Conv1D(32,kernel_size=100,activation='relu'))
model.add(tf.keras.layers.MaxPool1D(pool_size=5))
model.add(tf.keras.layers.Dropout(0.6))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(1,activation='linear'))
model.compile(loss='mse',optimizer='ADAM',metrics=['mae',RSquare()])
model.summary()
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```
